```python
def anagram(string1,string2):
    a = string1.replace(" ","").lower()
    b = string2.replace(" ","").lower()
    
    return sorted(a) == sorted(b)
```

```python
def anagram2(string1,string2):
    s1 = string1.replace(" ","").lower()
    s2 = string2.replace(" ","").lower()
    
    if len(s1) != len(s2):
        return False
    
    count = {}
    for letter in s1:
        if letter in count:
            count[letter]+=1
        else:
            count[letter] = 1
    for letter in s2:
        if letter in count:
            count[letter]-=1
        else:
            count[letter] = 1
    for k in count:
        if count[k]!=0:
            return False
        return True
    
```

```python
anagram2("shivam","lola")
```

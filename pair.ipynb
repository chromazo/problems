```python
def pair(arr,k):
    
    #edge case
    if len(arr)<2:
        print("Error")
    
    #using sets for tracking
    seen = set()
    output = set()
    
    for num in arr:
        diff = k-num
        
        if diff not in seen:
            seen.add(num)
        
        else:
            output.add((min(num,diff),max(num,diff)))
    
#     return(len(output))
    print("\n".join(map(str,list(output))))
```

```python
pair([1,3,2,2],4)
```
